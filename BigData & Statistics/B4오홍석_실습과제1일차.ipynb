### 실습문제 4

- page 34

```python
np.random.seed(1234)

df = pd.DataFrame(randn(10,4), columns = ['A','B','C', 'D'])
```

```python
## 2-1. B열
display(df[['B']])

## 2-2. B,D열
display(df[['B', 'D']])

## 2-3. 4~8행
display(df[4:9])
```

```python
## 2-1-loc. B열
display(df.loc[2:8, ['A','C']])

## 2-2-loc. B,D열
display(df.loc[:, ['A','C']])

## 2-3-loc. 3~8행
display(df.loc[3:8])

## 2-4-loc. 3행
display((df.loc[3:3]))
```

```python
## 3-1-iloc. 0~6행, A열
display(df.iloc[0:7, [0]])

## 3-2-iloc. A열
display(df.iloc[:,[0]])

## 3-3-iloc. 2~6행
display(df.iloc[2:7, :])

## 3-4-iloc. 5행
display((df.iloc[5:6, :]))
```

```python
## 4-1-query. D > 0.5
display(df[df.D > 0.5])

## 4-2-query. B > D
display(df[df.B > df.D])

## 4-3-query. A > 0.5 & D < 0.0
display(df.query('A > 0.5 & D < 0.0'))

## 4-4-query. A > 0.5 | D < 0.0
display(df.query('A > 0.5 | D < 0.0'))
```
